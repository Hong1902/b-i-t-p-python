W4A6 Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
while True:
    try:
        # Đọc từng dòng, bẻ ra thành a và b
        line = input()
        if not line:
            break
        a, b = map(int, line.split())
        
        tong = 0 # Khởi tạo biến tổng nằm ngoài vòng lặp duyệt x
        
        for x in range(a, b + 1):
            if x > 1:
                y= True # Giả sử x ban đầu là số nguyên tố
                
                # Kiểm tra xem x có chia hết cho số nào từ 2 đến căn bậc 2 của nó không
                for i in range(2, int(x**0.5) + 1):
                    if x % i == 0:
                        y = False # Nếu có, x không phải số nguyên tố
                        break # Thoát vòng lặp con ngay lập tức để tối ưu code
                
                # Nếu duyệt xong mà is_prime vẫn là True thì cộng vào tổng
                if y:
                    tong += x
                    
        print(f'Tổng của các số nguyên tố từ {a} đến {b} là {tong}')
        
    except EOFError:
        # Khi hết dòng đọc (hết test case), thoát chương trình
        break
```

W4A8 Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())

so_buoc = 0 # Tạo biến để đếm xem đã cộng bao nhiêu lần

# Lấy số đảo ngược của n bằng tuyệt chiêu cắt chuỗi
dao_nguoc = int(str(n)[::-1])

# Chừng nào n CÒN KHÁC số đảo ngược (tức là chưa phải Palindrome)
while n != dao_nguoc:
    n = n + dao_nguoc             # Lấy n cộng với số đảo ngược
    dao_nguoc = int(str(n)[::-1]) # Cập nhật lại số đảo ngược mới cho kết quả n vừa cộng
    so_buoc += 1                  # Tăng số bước lên 1

# Vòng lặp chỉ dừng lại khi n == dao_nguoc. Lúc này ta in ra kết quả:
print(so_buoc, n)
```

W4A27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
# Nhập chuỗi từ bàn phím
chuoi = input("Nhập vào một chuỗi: ")

# Bước 1: Tính chiều dài của chuỗi (vì không được dùng len)
chieu_dai = 0
for ky_tu in chuoi:
    chieu_dai += 1

# Bước 2: Đếm số từ bằng vòng lặp for trên index
so_tu = 0
dang_o_trong_tu = False  # Biến trạng thái (Cờ) để biết ta có đang xét một từ hay không

for i in range(chieu_dai):
    if chuoi[i] != ' ':  # Nếu ký tự tại vị trí i là chữ cái
        if dang_o_trong_tu == False:
            # Đây là chữ cái đầu tiên của một từ mới
            so_tu += 1
            dang_o_trong_tu = True
    else:  # Nếu ký tự tại vị trí i là khoảng trắng
        dang_o_trong_tu = False # Đánh dấu là đã thoát ra khỏi từ

print(f"Số từ trong chuỗi là: {so_tu}")
```

W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa

```python
chuoi = input("Nhập vào chuỗi: ")
chuoi_moi = ""

for ky_tu in chuoi:
    if 'a' <= ky_tu <= 'z':
        # Đổi thành số ASCII, trừ đi 32 để thành chữ hoa, rồi ép ngược lại thành chữ
        chuoi_moi += chr(ord(ky_tu) - 32) 
    elif 'A' <= ky_tu <= 'Z':
        # Đổi thành số ASCII, cộng thêm 32 để thành chữ thường
        chuoi_moi += chr(ord(ky_tu) + 32)

print(chuoi_moi)
```

W4A31 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
chuoi = input("Nhập vào một chuỗi: ")

# Khởi tạo 3 biến đếm (giống như chuẩn bị 3 cái giỏ để phân loại)
dem_hoa = 0
dem_thuong = 0
dem_so = 0

# Duyệt qua từng ký tự trong chuỗi
for ky_tu in chuoi:
    # Nếu rơi vào khoảng từ 'A' đến 'Z' -> Bỏ vào giỏ In Hoa
    if 'A' <= ky_tu <= 'Z':
        dem_hoa += 1
        
    # Nếu rơi vào khoảng từ 'a' đến 'z' -> Bỏ vào giỏ In Thường
    elif 'a' <= ky_tu <= 'z':
        dem_thuong += 1
        
    # Nếu rơi vào khoảng từ '0' đến '9' -> Bỏ vào giỏ Số
    elif '0' <= ky_tu <= '9':
        dem_so += 1

# In kết quả ra màn hình
print(f"Số ký tự in hoa: {dem_hoa}")
print(f"Số ký tự in thường: {dem_thuong}")
print(f"Số ký tự số: {dem_so}")
```
