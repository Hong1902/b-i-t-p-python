Bài 4 :Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
# 1. In khoảng trắng ở đầu mỗi dòng để tạo độ nghiêng
    print(" " * (h - i), end=" ")
    
    # 2. In dấu sao
    if i == 1:
        print("*") # Dòng đỉnh
    elif i == h:
        print("*" * (2 * h - 1)) # Dòng đáy (có 2h-1 dấu sao)
    else:
        # Dòng giữa: 1 sao + khoảng trống giữa + 1 sao
        # Khoảng trống giữa tăng dần theo quy luật: 2*i - 3
        print("*" + " " * (2 * i - 3) + "*")
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
a = int(input("Nhập số: "))

# Nếu số đầu tiên nhập vào đã là -1 thì không có gì để tính toán
if a == -1:
    print("Chưa có số hợp lệ nào được nhập.")
else:
    # Khởi tạo max và min bằng chính số đầu tiên vừa nhập
    so_lon_nhat = a
    so_nho_nhat = a

    # Vòng lặp chạy chừng nào số vừa nhập không phải là -1
    while a != -1:
        # Kiểm tra xem số hiện tại có lớn hơn max hay nhỏ hơn min không
        if a > so_lon_nhat:
            so_lon_nhat = a
        if a < so_nho_nhat:
            so_nho_nhat = a
            
        # Cuối vòng lặp, yêu cầu nhập số tiếp theo
        a = int(input("Nhập số: "))

    # In kết quả sau khi đã nhập -1 và thoát vòng lặp
    print("Số lớn nhất là:", so_lon_nhat)
    print("Số nhỏ nhất là:", so_nho_nhat)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
A=int(input())
x=1
y=1
while x+y<=A:
    z=x+y
    x=y
    y=z
print(y)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input("Nhập vào số nguyên dương n: "))

temp = n
dem_nguyen_to = 0

# Lặp để bóc tách từng chữ số
while temp > 0:
    chu_so = temp % 10  # Lấy chữ số ở hàng đơn vị
    
    # Kiểm tra xem chữ số vừa lấy có phải là 2, 3, 5 hoặc 7 không
    if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
        dem_nguyen_to += 1
        
    temp = temp // 10  # Cắt bỏ chữ số cuối cùng đi

print(f"Số lượng chữ số là số nguyên tố trong {n} là: {dem_nguyen_to}")
```
