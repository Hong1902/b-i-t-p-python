Cho trước N ngày, hãy đổi N thành số năm, số tuần và số ngày. Biết rằng một năm có 365 ngày.

```python
N=int(input())
if N>=365:
    a=N//365
    b=(N-365*a)//7
    c=(N-365*a)%7
    print(a,b,c)
elif 7<=N<365:
    a=0
    b=N//7
    c=N%7
    print(a,b,c)
elif 0<=N<7:
    a=0
    b=0
    c=N
    print(a,b,c)
else:
    print("Vô lý")

```
