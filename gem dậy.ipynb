```python
n=int(input())
x=1
for i in range(1,n+1):
    x=x*i
print(x)
```

Nhập vào một số nguyên dương (ví dụ: 123). Tính tổng các chữ số của nó ($1+2+3=6$).

```python
n=int(input())
x=n
y=0
while x>0:
    y=x%10+y
    x=x//10
print(y)
```

 In ra $n$ số đầu tiên trong dãy Fibonacci (0, 1, 1, 2, 3, 5, 8...).

```python
n = int(input("Bạn muốn in bao nhiêu số Fibonacci? "))
count = 0
a=0
b=1
while count < n:
    print(a, end=" ")
    c=a
    a=b
    b=c+b
    count += 1
```

Vẽ hình thoi bằng dấu sao (*)

```python
n=int(input())
for i in range(1,n+1):
  print(" "*(n-i),end=" ")
  if i==1:
    print("*")
  else:
    print("*"*(2*i-1))
for x in range(n-1,0,-1):
  print(" "*(n-x),end=" ")
  if x==1:
    print("*")
  else:
    print("*"*(2*x-1))
```

Khi bạn viết code cho một hệ thống bảo mật (như khóa cửa thông minh hay thẻ ATM), bạn phải dùng vòng lặp vô hạn để yêu cầu người dùng nhập mật khẩu. Nhưng nếu nhập sai quá 3 lần thì phải khóa hệ thống (dùng lệnh break).

```python
mat_khau=190207
so_lan_thu_con_lai=3
while True:
    n=int(input())
    if n==mat_khau:
        print("Mở khóa thành công")
        break
    else:
        so_lan_thu_con_lai=so_lan_thu_con_lai-1
    print(f"so_lan_thu_con_lai={so_lan_thu_con_lai}")
    if so_lan_thu_con_lai==0:
        print("bạn đã bị khóa")
        break
```
