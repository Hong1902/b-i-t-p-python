W7A16 Cho một mảng số nguyên không âm. Gọi
mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y
là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp.


```python
n=list(map(int,input().split()))
d={}
y=0
for i in n:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1
for keys in d.keys():
    if int(keys)==int(d[keys]):
        y=y
    elif int(keys)>int(d[keys]):
        y=y+int(d[keys])
    else:
        y=y+(int(d[keys])-int(keys))
if y==0:
    print("Yes")
else:
    print("No",y)

    
    
```

W7A17 Tìm phần tử xuất hiện nhiều hơn n/2 lần
trong mảng (nếu có) với độ phức tạp thời
gian là O(n), độ phức tạp bộ nhớ là O(1),
- Boyer–Moore. Nếu không có phần tử
nào thỏa mãn thì không in gì


```python
n = list(map(int, input().split()))
length = len(n)
candidate = None
count = 0
for x in n:
    if count == 0:
        candidate = x
        count = 1
    elif x == candidate:
        count += 1
    else:
        count -= 1
if candidate is not None:
    actual_count = 0
    for x in n:
        if x == candidate:
            actual_count += 1
    if actual_count > length / 2:
        print(candidate)
```

nhiều tên cùng số lần gặp thì lấy tên có

W7A27 Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra
xem dãy ngoặc đã cho có hợp lệ hay
không


```python
def check_valid_brackets(s):
    stack = []
    bracket_map = {')': '(', ']': '[', '}': '{'}
    for char in s:
        if char in bracket_map.values(): 
            stack.append(char)
        elif char in bracket_map.keys(): 
            if not stack or stack.pop() != bracket_map[char]:
                return False
        else:
            continue
    return len(stack) == 0
n=input()
print(check_valid_brackets(n))
```

W7A28 Bà̀i toán Tìm dãy các hợp số dài nhất liên
tiếp không chứa số nguyên nào <= N (N
là số nguyên dương cho trước). Nếu có
nhiều dãy cùng độ dài thì lấy dãy nhỏ
nhất (gặp đầu tiên từ trái qua)

```python
def find_longest_composite_sequence(n):
    if n < 4:
        return None
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    for p in range(2, int(n**0.5) + 1):
        if is_prime[p]:
            for i in range(p * p, n + 1, p):
                is_prime[i] = False
    check_points = [1] 
    for i in range(2, n + 1):
        if is_prime[i]:
            check_points.append(i)
    check_points.append(n + 1)
    max_len = 0
    start_num = -1
    for i in range(len(check_points) - 1):
        p1 = check_points[i]
        p2 = check_points[i+1]
        current_len = p2 - p1 - 1
        
        if current_len > max_len:
            max_len = current_len
            start_num = p1 + 1
    if max_len > 0:
        result = [str(x) for x in range(start_num, start_num + max_len)]
        print(" ".join(result))
    else:
        print("Không tìm thấy dãy hợp số")

n=int(input())
find_longest_composite_sequence(n)
```

W7A29 Tìm Dãy con tăng dần liên tục dài nhất
của một danh sách nhập vào từ bàn
phím. Lấy dãy đầu tiên nếu có nhiều dãy
có số phần tử bằng nhau

```python
def find_longest_increasing_subsequence():
    try:
        data = list(map(int, input().split()))
    except EOFError:
        return
    if not data:
        return
    max_len = 1
    best_start = 0
    current_len = 1
    current_start = 0
    for i in range(1, len(data)):
        if data[i] > data[i-1]:
            current_len += 1
        else:
            if current_len > max_len:
                max_len = current_len
                best_start = current_start
            current_len = 1
            current_start = i
    if current_len > max_len:
        max_len = current_len
        best_start = current_start
    result = data[best_start : best_start + max_len]
    print(*(result))
find_longest_increasing_subsequence()
```

W7A30 Tìm Dãy con tăng dần KHÔNG liên tục
dài nhất của một danh sách nhập vào từ
bàn phím 

```python
def find_lis():
    try:
        line = input().split()
        if not line: return
        a = [int(x) for x in line]
    except EOFError:
        return
    n = len(a)
    L = [1] * n
    trace = [-1] * n
    for i in range(1, n):
        for j in range(i):
            if a[j] < a[i]:
                if L[j] + 1 > L[i]:
                    L[i] = L[j] + 1
                    trace[i] = j
    max_len = max(L)
    end_index = L.index(max_len) 
    lis = []
    while end_index != -1:
        lis.append(a[end_index])
        end_index = trace[end_index]
    lis.reverse()
    print(*(lis))
find_lis()
```

W7A31 Đổi chỗ đúng 2 chữ số để được số nhỏ
nhất có thể

```python
def solve():
    s = input().strip()
    a = list(s)
    n = len(a)
    for i in range(n - 1):
        min_idx = -1
        for j in range(n - 1, i, -1):
            if i == 0 and a[j] == '0': continue 
            if a[j] < a[i]:
                if min_idx == -1 or a[j] < a[min_idx]:
                    min_idx = j
        if min_idx != -1:
            a[i], a[min_idx] = a[min_idx], a[i]
            print("".join(a))
            return
    if n >= 2:
        a[n-2], a[n-1] = a[n-1], a[n-2]
        print("".join(a))
    else:
        print(s) 

solve()
```
