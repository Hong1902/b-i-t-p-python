W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def tim_vi_tri_don_gian(danh_sach, k):
    so_luong = len(danh_sach)
    for i in range(so_luong):
        if danh_sach[i] == k:
            return i+1
    return -1
print("--- BƯỚC 1: TẠO DANH SÁCH ---")
danh_sach_nguoi_dung = []
so_luong_nhap = int(input("Danh sách của bạn có bao nhiêu số? "))
for i in range(so_luong_nhap):
    so = int(input(f"Nhập số ở vị trí {i}: "))
    danh_sach_nguoi_dung.append(so)
print("\n--- BƯỚC 2: NHẬP SỐ CẦN TÌM ---")
so_k = int(input("Nhập số nguyên dương k mà bạn muốn tìm: "))
print("\n--- KẾT QUẢ ---")
print(f"Danh sách bạn vừa nhập là: {danh_sach_nguoi_dung}")
ket_qua = tim_vi_tri_don_gian(danh_sach_nguoi_dung, so_k)
if ket_qua != -1:
    print(f"-> Tìm thấy! Số {so_k} xuất hiện lần đầu tiên ở vị trí thứ: {ket_qua}")
else:
    print(f"-> Không tìm thấy số {so_k} trong danh sách. Kết quả: {ket_qua}")
```

W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def tinh_khoang_cach_hamming(x, y):
    return bin(x ^ y).count('1')
print("--- CHƯƠNG TRÌNH TÍNH KHOẢNG CÁCH HAMMING ---")
try:
    so_thu_nhat = int(input("Nhập số nguyên thứ nhất (x): "))
    so_thu_hai = int(input("Nhập số nguyên thứ hai (y): "))
    khoang_cach = tinh_khoang_cach_hamming(so_thu_nhat, so_thu_hai)
    print("\n--- KẾT QUẢ ---")
    print(f"Khoảng cách Hamming giữa {so_thu_nhat} và {so_thu_hai} là: {khoang_cach}")
    nhi_phan_x = bin(so_thu_nhat)[2:] 
    nhi_phan_y = bin(so_thu_hai)[2:]
    do_dai_max = max(len(nhi_phan_x), len(nhi_phan_y))
    nhi_phan_x = nhi_phan_x.zfill(do_dai_max)
    nhi_phan_y = nhi_phan_y.zfill(do_dai_max)
    print("\nGiải thích chi tiết bằng nhị phân:")
    print(f"Số x ({so_thu_nhat:3}):  {nhi_phan_x}")
    print(f"Số y ({so_thu_hai:3}):  {nhi_phan_y}")
    print(f"Có tổng cộng {khoang_cach} vị trí bit khác nhau giữa hai hàng trên.")
except ValueError:
    print("\n[Lỗi] Bạn phải nhập vào một số nguyên hợp lệ!")
```

W5A10 Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 

```python
def kiem_tra_dang_cau(a, b):
    map_a_to_b = {}
    map_b_to_a = {}
    for char_a, char_b in zip(a, b):
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a           
    return True
print("=== CHƯƠNG TRÌNH KIỂM TRA CHUỖI ĐẲNG CẤU ===")
print("Lưu ý: Đề bài yêu cầu hai chuỗi phải có cùng độ dài.\n")
tu_1 = input("Nhập chuỗi thứ nhất (a): ")
tu_2 = input("Nhập chuỗi thứ hai (b): ")
print("-" * 40)
if len(tu_1) != len(tu_2):
    print("[Lỗi]: Hai chuỗi bạn nhập có độ dài khác nhau!")
    print(f"   Chuỗi '{tu_1}' dài {len(tu_1)} ký tự.")
    print(f"   Chuỗi '{tu_2}' dài {len(tu_2)} ký tự.")
    print("Vui lòng chạy lại chương trình và nhập hai chuỗi dài bằng nhau.")
else:
    ket_qua = kiem_tra_dang_cau(tu_1, tu_2)
    if ket_qua == True:
        print(f" Kết quả: True")
        print(f"   Giải thích: Hai chữ '{tu_1}' và '{tu_2}' LÀ chuỗi đẳng cấu.")
    else:
        print(f" Kết quả: False")
        print(f"   Giải thích: Hai chữ '{tu_1}' và '{tu_2}' KHÔNG PHẢI là chuỗi đẳng cấu.")
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
a,b=map(int,input().split())
op=input()
def compute_hien_dai(a, b, op):
    match op:
        case '+':
            return a + b
        case '-':
            return a - b
        case '*':
            return a * b
        case '/':
            return a / b if b != 0 else "Lỗi chia cho 0"
        case _:
            return "Phép toán không hợp lệ"
print(compute_hien_dai(a,b,op))
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
s = input("Nhập vào một giá trị: ")

def detect_number_type(s):
    # Xử lý trường hợp số âm: bỏ dấu trừ ở đầu nếu có
    s_check = s.lstrip('-')
    
    # 1. Kiểm tra số nguyên: Nếu chuỗi chỉ toàn là số
    if s_check.isdigit():
        return 'int'
    
    # 2. Kiểm tra số thực: Nếu xóa 1 dấu chấm đi mà nó là số
    elif '.' in s_check and s_check.replace('.', '', 1).isdigit():
        return 'float'
    
    # 3. Còn lại không phải số
    else:
        return 'neither'

print(detect_number_type(s))
```
