W6A1 Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
n=list(map(int,input().split()))
a=[]
for i in n:
    if i not in a:
        a.append(i)
print(a)
```

W6A4 Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó

```python
n=input().split()
result={}
for i in n:
    key,value=i.split(":")
    if key in result:
        result[key].append(value)
    else:
        result[key]=[value]
print(result)
    
```

W6A5 Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...} (đếm số >0, <0, =0)

```python
n=list(map(int,input().split()))
d={'positives':0,'negatives':0,'zero':0}
for i in n:
    if i>0:
        d['positives']+=1
    elif i<0:
        d['negatives']+=1
    else:
        d['zero']+=1
print(d)
```

W6A9 Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key

```python
n=input().split()
m=input().split()
d={}
for i in n:
    key,value=i.split(":")
    value=int(value)
    if key in d:
        d[key]+=value
    else:
        d[key]=value
for a in m:
    key,value=a.split(":")
    value=int(value)
    if key in d:
        d[key]+=value
    else:
        d[key]=value
sorted_d={}
for k in sorted(d.keys()):
    sorted_d[k]=d[k]
print(sorted_d)
    
```

W6A10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple.

```python
n=list(map(int,input().split()))
b=int(input())
a=len(n)
c=[]
for i in range(a):
    for k in range(i+1,a):
        if n[i]+n[k]==b:
            c.append((n[i],n[k]))
c.sort()
print(c)
            
    
```

W6A11 [Tách số chẵn và lẻ từ tuple]
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt. 


```python
n=tuple(map(int,input().split()))
chan=[]
le=[]
for i in n:
    if i%2==0:
        chan.append(i)
    else:
        le.append(i)
tuple_chan=tuple(chan)
tuple_le=tuple(le)
print(tuple_chan)
print(tuple_le)
```

W6A13 [Đảo ngược key và value của dictionary]

W6A14 [Giao của hai danh sách]

phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các

W6A15 [Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới

```python
n=input().split()
a=len(n)
k=int(input())
d={}
for i in range(1,a,2):
    n[i]=int(n[i])
    if n[i]>k:
        d[n[i-1]]=n[i]
print(d)
        
```
