W5A1 Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> 
int:

```python
def Max_Of_Two(a: int, b: int) -> int:
    """
    Hàm trả về số lớn hơn trong hai số nguyên a và b.
    """
    if a > b:
        return a
    else:
        return b
```
